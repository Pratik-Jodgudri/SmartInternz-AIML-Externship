```python
df['mpg'].plot()
df['hp'].plot()
plt.legend(['mpg', 'hp'])
```

```python
sns.displot(df['mpg'], bins=30, kde=True)
plt.xlim(0, 50)
plt.show()
```

```python
plt.pie([32, 15, 23, 43, 55], autopct='%.3f%%', colors=['r', 'y', 'g', 'b', 'k'])
plt.legend(['a', 'b', 'c','d','e'])
plt.show()
```

```python
plt.scatter(x=df['hp'], y=df['mpg'])
```

```python
x=df['mpg']
y=df['hp']
fig,ax = plt.subplots(2,2)
ax[0,0].plot([1, 2, 3, 4], [32, 55, 12, 79])
ax[0,1].pie([32, 15, 23, 43, 55])
ax[1,0].bar(x,y)
ax[1,1].scatter(x,y)
```
